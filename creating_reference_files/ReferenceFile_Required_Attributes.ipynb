Required Attributes for Reference Files
===============================

The following summarizes what attributes are needed for the various reference
file types. This information is obtained from the corresponding schema files.
Ideally this page is updated when those are, but the schema files are always
the original source. Since they can be a little difficult to interpret, this 
summary is to make it easier to understand.

All these reference files use the attributes shown in "Common" and all these
attributes appear in the meta attribute. Merge what is shown in meta for the
specific reference file with those shown in common.

Each attribute will usually list the type (unless the list of permitted 
values clearly indicate it is a string) along with other possible restrictions
on the allowed value (enumerated list of allowed values, min and/or max 
values)

Common Reference File Attributes
-----------------------------------------

All these attributes are under the meta attribute and all are required.
Indented lists are attributes under the attribute they are listed after.

- telescope: Value must be: ROMAN
- instrument
  - name: Value must be: WFI
  - detector: Value must be one of: WFI01, WFI02,..., WFI18
  - optical_element: Value must be one of: F062, F087, F106, F129,
    W146, F158, F184, GRISM, PRISM, DARK, or F213
- reftype: Value must be one of: AREA, DARK, FLAT, GAIN, IPC,
  LINEARITY, PERSAT, PHOTOM, READNOISE, REFPIX, SATURATION,
  SUPERBIAS, TRAPDENSITY, or TRAPPARS
- pedigree: Value must be one of GROUND, MODEL, DUMMY, or SIMULATION
- description: string
- author: string
- useafter: Astropy time.Time object

Dark
----

- meta
  - reftype: Value must be: DARK
- data: Value must be float32 numpy array with 3 dimensions
- dq: Value must be uint32 numpy array with 2 dimensions
- err: Value must be float32 numpy array with 3 dimensions

Flat
----

- meta
  - reftype: Value must be: FLAT
- data: Value must be float32 numpy array with 2 dimensions  
- dq: Value must be uint32 numpy array with 2 dimensions
- err: Value must be float32 numpy array with 2 dimensions

Gain
----

- meta
  - reftype: Value must be: GAIN
- data: Value must be float32 numpy array with 2 dimensions

Mask
----

- meta
  - reftype: Value must be: MASK
- dq: Value must be uint32 numpy array with 2 dimensions


Readnoise
------------

- meta
  - reftype: Value must be: READNOISE
- data: Value must be float32 numpy array with 2 dimensions.
